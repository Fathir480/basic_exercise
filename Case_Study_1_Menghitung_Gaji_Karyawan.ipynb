## **Case Study 1: Menghitung Gaji Karyawan**
___


Pak Bakir adalah seorang manager di sebuah Perusahaan B. Beliau ingin menghitung gaji karyawannya selama satu bulan secara keseluruhan. Daftar karyawan Pak Bakir ada di bawah ini.

|No|Nama|Jam Kerja|Jam Lembur | Upah Kerja (Kotor) | Upah Lembur | Total Upah | Pajak | Total Upah (Netto)
|----|---|---|---|---|---|---|---|---|
|1|Joko|44|12|
|2|Alex | 48 | 11 |
|3|Chandra | 50 | 10 |
|4|Hartono | 42 | 15 |
|5|Samsul | 45 | 12 |

Soal: 
- Hitunglah Upah Kerja (Kotor) dari seluruh karyawan Perusahaan B jika diketahui upah per jam adalah Rp 30000
- Hitunglah Upah Lembur dari seluruh karyawan Perusahaan B
- Hitunglah Total Upah dari seluruh karyawan Perusahaan B
- Jika persentase pajak karyawan adalah 5%, hitunglah Pajak dari setiap karyawan Perusahaan B
- Hitunglah Total Upah (Netto) dari setiap karyawan Perusahaan B

```python
# Inisialisasi variable
UPAH_PER_JAM = 30000
persentase_pajak = 0.05

# Joko
nama_karyawan_1 = "Joko"
jam_kerja_karyawan_1 = 44
jam_lembur_karyawan_1 = 12

# Alex
nama_karyawan_2 = "Alex"
jam_kerja_karyawan_2 = 48
jam_lembur_karyawan_2 = 11

# Chandra
nama_karyawan_3 = "Chandra"
jam_kerja_karyawan_3 = 50
jam_lembur_karyawan_3 = 10

# Hartono
nama_karyawan_4 = "Hartono"
jam_kerja_karyawan_4 = 42
jam_lembur_karyawan_4 = 15

# Samsul
nama_karyawan_5 = "Samsul"
jam_kerja_karyawan_5 = 45
jam_lembur_karyawan_5 = 12
```

```python
# Hitung Upah Kerja
upah_kerja_karyawan_1 = UPAH_PER_JAM * jam_kerja_karyawan_1
upah_kerja_karyawan_2 = UPAH_PER_JAM * jam_kerja_karyawan_2
upah_kerja_karyawan_3 = UPAH_PER_JAM * jam_kerja_karyawan_3
upah_kerja_karyawan_4 = UPAH_PER_JAM * jam_kerja_karyawan_4
upah_kerja_karyawan_5 = UPAH_PER_JAM * jam_kerja_karyawan_5

# Hitung Upah Lembur
upah_lembur_karyawan_1 = UPAH_PER_JAM * jam_lembur_karyawan_1
upah_lembur_karyawan_2 = UPAH_PER_JAM * jam_lembur_karyawan_2
upah_lembur_karyawan_3 = UPAH_PER_JAM * jam_lembur_karyawan_3
upah_lembur_karyawan_4 = UPAH_PER_JAM * jam_lembur_karyawan_4
upah_lembur_karyawan_5 = UPAH_PER_JAM * jam_lembur_karyawan_5

# Hitung Total Upah 
total_upah_karyawan_1 = upah_kerja_karyawan_1 + upah_lembur_karyawan_1
total_upah_karyawan_2 = upah_kerja_karyawan_2 + upah_lembur_karyawan_2
total_upah_karyawan_3 = upah_kerja_karyawan_3 + upah_lembur_karyawan_3
total_upah_karyawan_4 = upah_kerja_karyawan_4 + upah_lembur_karyawan_4
total_upah_karyawan_5 = upah_kerja_karyawan_5 + upah_lembur_karyawan_5

# Hitung pajak karyawan
pajak_karyawan_1 = persentase_pajak * total_upah_karyawan_1
pajak_karyawan_2 = persentase_pajak * total_upah_karyawan_2
pajak_karyawan_3 = persentase_pajak * total_upah_karyawan_3
pajak_karyawan_4 = persentase_pajak * total_upah_karyawan_4
pajak_karyawan_5 = persentase_pajak * total_upah_karyawan_5

# Hitung Upah bersih karyawan
upah_karyawan_1 = total_upah_karyawan_1 - pajak_karyawan_1
upah_karyawan_2 = total_upah_karyawan_2 - pajak_karyawan_2
upah_karyawan_3 = total_upah_karyawan_3 - pajak_karyawan_3
upah_karyawan_4 = total_upah_karyawan_4 - pajak_karyawan_4
upah_karyawan_5 = total_upah_karyawan_5 - pajak_karyawan_5
```

```python
# Tampilan 

# Joko
print(f"""Upah kerja Joko = {upah_kerja_karyawan_1},
Upah lembur Joko = {upah_lembur_karyawan_1},
Total upah Joko = {total_upah_karyawan_1},
Pajak Joko = {pajak_karyawan_1},
Upah bersih Joko = {upah_karyawan_1}
""")

# Alex
print(f"""Upah kerja Alex = {upah_kerja_karyawan_2},
Upah lembur Alex = {upah_lembur_karyawan_2},
Total upah Alex = {total_upah_karyawan_2},
Pajak Alex = {pajak_karyawan_2},
Upah bersih Alex = {upah_karyawan_2}
""")

# Chandra
print(f"""Upah kerja Chandra = {upah_kerja_karyawan_3},
Upah lembur Chandra = {upah_lembur_karyawan_3},
Total upah Chandra = {total_upah_karyawan_3},
Pajak Chandra = {pajak_karyawan_3},
Upah bersih Chandra = {upah_karyawan_3}
""")

# Hartono
print(f"""Upah kerja Hartono = {upah_kerja_karyawan_4},
Upah lembur Hartono = {upah_lembur_karyawan_4},
Total upah Hartono = {total_upah_karyawan_4},
Pajak Hartono = {pajak_karyawan_4},
Upah bersih Hartono = {upah_karyawan_4}
""")

# Samsul
print(f"""Upah kerja Samsul = {upah_kerja_karyawan_5},
Upah lembur Samsul = {upah_lembur_karyawan_5},
Total upah Samsul = {total_upah_karyawan_5},
Pajak Samsul = {pajak_karyawan_5},
Upah bersih Samsul = {upah_karyawan_5}
""")
```
